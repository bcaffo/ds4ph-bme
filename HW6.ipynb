```
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

x = np.array([1.57, 1.25, 2.80, 0.43])
w = np.array([2,2,1,1])

mu = sum(w*x)/sum(w)

print('The optimal value of mu is ' + str(mu))
```

```
data = pd.read_csv('shhs1.txt', sep = '\t')
data['log_rdi4p'] = np.log(data['rdi4p']+1)
data1 = data[['bmi_s1', 'log_rdi4p']]
data2 = data1.dropna()
datax = data2[['bmi_s1']]
datay = data2[['log_rdi4p']]

reg = LinearRegression().fit(datax,datay)
m = reg.coef_
b = reg.intercept_

line_values_x = np.linspace(min(datax.values)[0],max(datax.values)[0],2)
line_values_y = m*line_values_x + b

plt.scatter(datax,datay)
plt.plot(line_values_x,line_values_y[0],color = 'r')
plt.show()
```

```
print('y-intercept is: ' + str(b[0]))
print('coefficient is: ' + str(m[0][0]))
print('predicted rdi4p with bmi = 30 is: ' + str(np.e**(30*m[0][0] + b[0])-1))
```
